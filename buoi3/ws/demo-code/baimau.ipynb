## Bài làm:

### Import thư viện cần dùng cho trực quan hóa dữ liệu

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Chuẩn bị dữ liệu:

```python
df = pd.read_csv("../data/HousingPrices-Amsterdam-August-2021.csv")
df = df[['Zip','Price','Area','Room']]
df['PricePerSqm'] = df['Price']/df['Area']
df.head(5)
```

```python
df_sort = df.sort_values('Price', ascending=False)
df_sort.head(5)
```

### Biểu đồ cột của 10 căn nhà có giá cao nhất trong database:

Basic

```python
plt.figure(figsize= (10,6))
x = df_sort['Zip'][0:10]
y = df_sort['Price'][0:10]
plt.bar(x,y)
plt.show()
```

Advance 1

```python
plt.figure(figsize=(10,6))
plt.bar(x,y)
plt.title("Top 10 khu vực có giá nhà cao nhất: ")
plt.xlabel('Zip Code', fontsize=12)
plt.xticks(fontsize=10)
plt.ylabel('House prices in million',fontsize=12)
plt.yticks(fontsize=10)
plt.show()
```

Advance 2

```python
fig, ax = plt.subplots(figsize=(50,18))
x = df_sort['Zip'][0:10]
y = df_sort['Price'][0:10]
y1 = df_sort['PricePerSqm'][0:10]
plt.subplot(1,2,1) #1 hàng 2 cột vị trí 1
plt.bar(x,y)
plt.xticks(fontsize=17)
plt.ylabel('House prices in millions', fontsize=25)
plt.yticks(fontsize=20)
plt.title('Top 10 khu vực có giá nhà cao nhất: ', fontsize=25)
plt.subplot(1,2,2)
plt.bar(x,y1)
plt.xticks(fontsize=17)
plt.ylabel('House prices per sqm', fontsize=25)
plt.yticks(fontsize=20)
plt.title('TTop 10 khu vực có giá nhà cao nhất mỗi mét vuông: ',fontsize=25)
plt.show()
```

### Sử dụng Seaborn

Basic

```python
plt.figure(figsize= (12,6))
data = df_sort[0:10]
sns.barplot(data= data, x= 'Zip',y = 'Price')
```

Advance 1

```python
plt.figure(figsize= (12,6))
data = df_sort[0:10]
ax = sns.barplot(data= data, x= 'Zip',y = 'Price')
ax.set_xlabel('Zip code',fontsize = 15)
ax.set_ylabel('House prices in millions', fontsize = 15)
ax.set_title('Top 10 Areas with the highest house prices', fontsize= 20)
```

Advance 2

```python
fig, ax = plt.subplots(1, 2,figsize=(40,18))
data = df_sort[0:10]
sns.set(font_scale = 3)
ax1 = sns.barplot(data= data, x= 'Zip',y = 'Price', ax = ax[0])
ax1.set_xlabel('Zip code')
ax1.set_ylabel('House prices in millions')
ax1.set_title('Top 10 Areas with the highest house prices')
ax2 = sns.barplot(data= data, x= 'Zip',y = 'PricePerSqm',ax=ax[1])
ax2.set_xlabel('Zip code')
ax2.set_ylabel('House prices per sqm')
ax2.set_title('Top 10 Areas with the highest price per sqm')
```
